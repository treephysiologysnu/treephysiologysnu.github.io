# 깃허브 사용 방법 정리
아래와 같은 명령어로 우선 아이디와 이메일을 설정하고,
~~~sh
git config --global user.name "User ID"
git config --global user.email "이메일 주소"
~~~

특정 폴더로 가서 README.md 파일이나 .gitignore 파일 생성
~~~sh
git init
touch README.md #touch .gitignore
~~~

지켜볼 파일을 추가하기. 슬래시(/)를 붙이면 그 아래의 모든 파일들을 포함시키게 된다.
~~~sh
git add README.md
git add some/dir/
~~~

추가한 파일들에 대한 commit을 할 차례:
~~~sh
git commit -m "First commit"
git push origin master
~~~

### SSH-Keys
`ssh-keygen` 으로 키 만든 뒤에 SSH-Key 추가하고([참고](#Annotation-1)) `~/.ssh` 경로에 들어가면 `id_rsa`, `id_rsa.pub` 이나 혹은 `known-hosts` 이런 식으로 파일이 생성되어 있음.

아래와 같은 명령어로 파일을 확인하고 깃허브에 등록해두기([참고](#Annotation-2))
~~~sh
cd ~/.ssh
cat id_rsa
~~~

그 뒤에도 계속 에러가 떠서, 검색해본 결과 아래와 같이 해야 작동한다는 이야기가 있었다.
~~~sh
git remote add origin git@github.com:treephysiologysnu/phenocam.tk-dev.git #NOT WORKING
git remote add origin https://github.com/treephysiologysnu/phenocam.tk-dev #WORKING
~~~

remote origin 지우는 방법은 다음과 같은 명령어로!
~~~sh
git remote rm origin
~~~

### Git commit 삭제하기!
만약 잘못된 commit이 되거나, 폴더 혹은 파일들이 add 된 경우 commit을 취소해야 한다. 아래와 같이 `git reset HEAD [file]` 명령어로 git add를 취소할 수 있다.
~~~sh
# Reverting Git add
git reset HEAD [file]
~~~

git commit 을 취소하는 경우에는 아래와 같이
~~~sh
git log # 로그 조회
git reset HEAD~2
~~~

### 최종 Push 단계
~~~sh
[Git] fatal: The remote end hung up unexpectedly
~~~
이라고 에러가 뜨게 되면 아래와 같은 명령어로 고치면 된단다.
~~~sh
git config --global http.postBuffer 20971520
~~~
*20971520 = 20 x 1024 x 1024 (20MB)*
