### 부트스트랩 적용

#### Overriding
부트스트랩 css 파일을 override 하기 위해서 id를 하나씩 새로 만들어주고, `phenocam-custom.css` 파일에다가 #을 붙여서 작성.
<br>처음에 이렇게 했는데 잘 안되서 아래와 같이 진행.
~~~css
#override-nav > ul > li > a:hover {
	color: #4285f4;
	background-color: #CCC;
}
~~~

이렇게 바꿨더니 잘 되더라.
~~~css
#override-nav > li > a:hover {
	color: #4285f4;
	background-color: #CCC;
}
~~~
